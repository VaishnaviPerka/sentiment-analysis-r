## 3.2 Tokenization
### 3.2.1 What is Tokenization?

Tokenization is the process of breaking a text into smaller units, called **tokens**. These tokens can be words, phrases, or even characters, depending on the level of tokenization.

#### Example:

#### Original Sentence:
*"I love data science!"*

#### Word Tokenization (Splitting by words):
`["I", "love", "data", "science", "!"]`

#### Character Tokenization (Splitting by each character):
`["I", " ", "l", "o", "v", "e", " ", "d", "a", "t", "a", " ", "s", "c", "i", "e", "n", "c", "e", "!"]`

#### Why is Tokenization Important?
- It helps in **text preprocessing** for machine learning models.
- It allows better **text analysis** by breaking large text into manageable parts.
- It is used in **sentiment analysis, text classification, and chatbots**.



### 3.2.2 Lets Do it practically
#### Creating a Text Data Frame

```r
# Create a sample dataframe
text_data <- data.frame(
  id = 1:6, 
  text = c(
   "The product broke within a week of use, not worth the money",
      "I am very disappointed with the poor quality and slow customer support.",
      "The product is good, but the quality is a bit poor, and the price is fair.",
      "Amazing product! The quality exceeded my expectations, highly recommend.",
      "Fantastic customer service, quick delivery, and the product works perfectly!",
      "The item didn’t match the description, and it was defective when I received it."
  )
)

text_data
```

#### Generating Tokens

**`%>%` (Pipe Operator)**: Passes the output of one function to the next, making code cleaner.  
**`unnest_tokens(word, text)`**: Breaks text into individual words for analysis. 

```r
# Word Tokenization using unnest_tokens
word_tokens <- text_data %>% unnest_tokens(word, text)
```

```r
head(word_tokens)
```

## 3.3 Remove Stop Words(Optional)

### 3.3.1 What are Stop Words?

In Natural Language Processing (NLP), **stop words** are words that are considered to have little to no meaningful value when analyzing text. These words are usually removed from the text during preprocessing because they don't contribute much to the analysis. Examples of stop words include words like **"the"**, **"is"**, **"in"**, **"and"**, **"to"**, etc.

Stop words are often common words used in everyday language that don't provide significant information in tasks like text classification, sentiment analysis, or topic modeling.

#### Why Remove Stop Words?

- **Reduces Noise**: Stop words can be noisy data that don't help the model.
- **Improves Model Efficiency**: By removing irrelevant words, we reduce the complexity of the data and help improve the performance of machine learning models.
- **Saves Processing Time**: Less data to process means faster analysis.

---

#### Example of Stop Words in a Sentence:
#### Original Sentence:
**"I love programming and learning new things every day."**

#### After Removing Stop Words:
**"love programming learning new things day."**

In this example, words like **"I"**, **"and"**, **"every"**, and **"to"** are removed because they don't add much meaning to the sentence for analysis.


### 3.3.2 anti_join()
 **`anti_join()`**: This function remove all the **stop words** from the **word_tokens**.

```r
words <- word_tokens %>% anti_join(stop_words, by= 'word')
```

```r
words
```

## 3.4 Sentiment Analysis

### 3.4.1 What is Sentiment Analysis?
Sentiment Analysis is used to determine the emotional tone of a piece of text. It helps classify text as **positive, negative, or neutral** based on the words used.  

#### Why is Sentiment Analysis Important?
- **Customer Feedback:** Businesses analyze product reviews and social media comments to understand customer satisfaction.  
- **Brand Reputation:** Companies monitor online discussions to see how people feel about their brand.  
- **Market Analysis:** Helps track public opinion on political issues, stock markets, and trends.  

#### Example of Sentiment Analysis
 
![Alt Text](images\sentimentanalysis.jpg) 

#### Types of Sentiment Lexicons
A **sentiment lexicon** is a collection of words labeled with sentiment scores. There are different lexicons available, such as:  
1. **AFINN** – Assigns words a sentiment score between -5 (very negative) and +5 (very positive).  
2. **Bing Liu's Lexicon** – Classifies words as either positive or negative.  
3. **NRC Lexicon** – Categorizes words based on emotions like anger, joy, fear, and surprise.  

For this analysis, we are using the **AFINN lexicon** because it provides a numerical sentiment score, making it easier to measure the intensity of sentiment in a sentence.  


#### How AFINN Lexicon Works
The **AFINN lexicon** assigns sentiment scores to words:  
- **Positive words** have **positive scores** (e.g., "great" = +3, "love" = +3).  
- **Negative words** have **negative scores** (e.g., "bad" = -2, "worst" = -3).  
- The total sentiment score of a sentence is calculated by adding up the scores of the words it contains.  

#### Example Calculation
Sentence: *"This is a great product but the service was bad."*  
- "great" = **+3**  
- "bad" = **-2**  
- **Total Sentiment Score = +1** (Overall Positive)  

By using the **AFINN lexicon**, we can efficiently analyze text sentiment based on numerical scores!   


### 3.4.2 Load the Afinn Lexicon

```r
afinn_lexicon <- read.csv('Data//Afinn.csv')
```

```r
#display 
head(afinn_lexicon)
```

### 3.4.3 inner_join()

**`inner_join()`**: This function returns rows where there is a match between the two data frames (afinn_lexicon and words) in the `"word"` column.

```r
afinn_sentiments <- words %>%
  inner_join(afinn_lexicon, by = "word")
```

```r
head(afinn_sentiments)
```

### 3.4.4 Calculating Score


- **`group_by()`**: Groups the data by the specific column.
- **`summarise(...)`**:
  - **`sentiment_words = paste(word, collapse = ", ")`**: Combines all words for each `id` into a single comma-separated string.
  - **`final_score = sum(value, na.rm = TRUE)`**: Sums the sentiment scores (`value`) for each `id`, ignoring any `NA` values.

The result is a new data frame `score` that shows, for each `id`, a list of words and the total sentiment score.

```r
score <- afinn_sentiments %>%
  group_by(id) %>% #group by id
  summarise(
    sentiment_words = paste(word, collapse = ", "),  # Combine words used for each id
    final_score = sum(value, na.rm = TRUE)  # Sum sentiment values
  )
```

```r
score
```

### 3.4.5 Classification

 Classifying the final_score into positive, negative, and neutral sentiments into a new column called `sentiment`

```r
score$sentiment <- ifelse(score$final_score > 0, "positive",
                   ifelse(score$final_score < 0, "negative", "neutral"))
```

```r
score
```

### 3.5.2 Data Cleaning

#### Removing Emojis 
The `remove_emojis` function removes emojis from a given text and replaces with an empty space.

- `gsub` function to find and replace emojis in the text.  
- `[\U0001F300-\U0001F6FF\U0001F900-\U0001F9FF\U0001FA70-\U0001FAFF\U00002702-\U000027B0\U000024C2-\U0001F251] ` - all these are the codes for wide range of emoji characters.

```r
remove_emojis <- function(text) {
  gsub("[\U0001F300-\U0001F6FF\U0001F900-\U0001F9FF\U0001FA70-\U0001FAFF\U00002702-\U000027B0\U000024C2-\U0001F251]", "", text, perl = TRUE)
}

#apply remove_emojis function to dataset
dataset$Online_comments <- remove_emojis(dataset$Online_comments)
```

```r
head(dataset)
```

#### Removing Foreign Languages

The `remove_non_english` function removes non-English characters, punctuation, and numbers from the text.

```r
remove_non_english <- function(text) {
  # Replace non-English characters with a space
  cleaned_text <- gsub("[^A-Za-z]", " ", text)

  # Remove extra spaces
  cleaned_text <- gsub("\\s+", " ", cleaned_text)

  return(cleaned_text)
}

#apply remove_non_english function to the dataset
dataset$Online_comments<- sapply(dataset$Online_comments, remove_non_english)
```
